```python
# change the wrong values to np.NaN
d.insulin.replace(0,np.nan, inplace=True)
d.triceps.replace(0,np.nan, inplace=True)
d.bmi.replace(0,np.nan,inplace = True)
```

```python
from sklearn.preprocessing import Imputer as imp
X = d.iloc[:,:-1]
y = d.iloc[:,-1]
print(X.shape,y.shape)
i=imp(missing_values = "NaN", strategy = "median",axis=0)
i.fit(X)
X = i.transform(X)

new_d = pd.DataFrame(X[:10])
print(new_d)
np.isnan(X).sum()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier

steps = [ ('i', imp(missing_values="NaN",strategy="mean",axis=0)),('clf', KNeighborsClassifier(n_neighbors = 6))]
X_train, X_test, y_train, y_test = t_t_s(X,y,test_size=0.13, random_state=32)

pipeline = Pipeline(steps)
pipeline.fit(X_train, y_train)
pipeline.predict(X_test)
pipeline.score(X_test, y_test)
```

```python
from sklearn.pipeline import Pipeline as pipe
from sklearn.neighbors import KNeighborsClassifier as kn

steps = [ ('imp', imp(missing_values="NaN",strategy="most_frequent",axis=0)),
          ('clf', kn(n_neighbors = 6))]


#split the data 
X_train, X_test, y_train, y_test = t_t_s(X,y,test_size=0.2, random_state=42)

pipeline = pipe(steps)

pipeline.fit(X_train, y_train)
pipeline.predict(X_test)
pipeline.score(X_test, y_test)
```
